```python
x = df.iloc[:, 1:11].values
y = df.iloc[:, 0].values
```

```python
import sklearn
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
lr=LinearRegression(normalize=True)
lr.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
lr_pred
```

```python
knn = KNeighborsClassifier(n_neighbors = 5, metric= 'minkowski', p=2)
knn.fit(x_train,y_train)
print('[1]K Neighbors Training Accuracy:', knn.score(x_train, y_train))
```

```python
tree = DecisionTreeClassifier(max_depth = 15, min_samples_leaf = 100)
tree.fit(x_train, y_train)    
```

```python
tree_accuracy=round(tree.score(x_train,y_train)*100)
tree_accuracy
```

```python
forest = RandomForestClassifier(n_estimators=400, max_depth=10, min_samples_leaf = 100,n_jobs=4)
forest.fit(x_train, y_train)

forest_pred=forest.predict(x_test)

forest_accuracy=round(forest.score(x_train,y_train)*100)
forest_accuracy
```
